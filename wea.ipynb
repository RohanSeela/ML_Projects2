```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
weather=pd.read_csv("weatherAUS.csv")
weather
```

```python
weather['RainToday'].replace({'No':0,'Yes':1}, inplace=True)
weather['RainTomorrow'].replace({'No':0,'Yes':1}, inplace=True)
```

```python
weather.corr()
```

```python
weather.apply(lambda x: sum(x.isnull()),axis=0)
```

```python
weather1 = weather[pd.notnull(weather['RainToday'])]
```

```python
weather1.apply(lambda x: sum(x.isnull()),axis=0)
```

```python
meanHumidity3pm=weather1["Humidity3pm"].mean()
meanHumidity3pm
```

```python
weather1["Humidity3pm"].fillna(meanHumidity3pm,inplace=True)
```

```python
medCloud3pm=weather1["Cloud3pm"].median()
medCloud3pm
```

```python
weather1["Cloud3pm"].fillna(medCloud3pm,inplace=True)
```

```python
weather1.apply(lambda x: sum(x.isnull()),axis=0)
```

```python
medgs=weather1["WindGustSpeed"].median()

weather1["WindGustSpeed"].fillna(medgs,inplace=True)
```

```python
weather["Pressure3pm"].value_counts()
```

```python

weather1["Pressure3pm"].fillna(1015.500,inplace=True)
```

```python
medt=weather1["Temp3pm"].median()

weather1["Temp3pm"].fillna(medt,inplace=True)
```

```python
wanted=weather1[["Rainfall","Sunshine","WindGustSpeed","Humidity3pm","Pressure3pm","Cloud3pm","Temp3pm","RainToday","RISK_MM","RainTomorrow"]]
```

```python
X=weather1[['Rainfall','WindGustSpeed','Humidity3pm','Pressure3pm','Cloud3pm','Temp3pm','RainToday']]
Y=weather1['RainTomorrow'] 
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.4,random_state=1)
```

```python
from sklearn.linear_model.logistic import LogisticRegression
lqsr=LogisticRegression(random_state=0)
lqsr.fit(X_train,Y_train) 
```

```python
X_test=X_test.reshape(-1,7)
```

```python
y_pred=lqsr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc_sco=accuracy_score(Y_test,y_pred)
acc_sco
```
